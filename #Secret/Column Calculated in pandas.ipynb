```python
import pandas as pd
import numpy as np
```

```python
data = {'nome': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 
        'idade': [42, 52, 36, 24, 73], 
        'preTeste': [4, 24, 31, 2, 3],
        'postTeste': [25, 94, 57, 62, 70]}
df = pd.DataFrame(data, columns = ['nome', 'idade', 'preTeste', 'postTeste'])
df
```

```python
df['sum'] = df['preTeste'] + df['postTeste']
df
```

```python
df['multi'] = df['preTeste'] * df['postTeste']
df
```

```python
df['avg_pre'] = df['preTeste']/2
df['avg_pos'] = df['postTeste']/2
df
```

```python
df['dif'] = df['postTeste'] - df['preTeste']
df
```

```python
df['%'] = df['preTeste']/df['postTeste']*100
df
```
