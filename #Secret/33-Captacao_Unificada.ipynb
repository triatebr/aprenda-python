```python
from unicodedata import normalize
def marca_detalhe(row):
    if (row['DS_MODALIDADE'] == 'UNIRONDON'):
       val = 'UNIC'
    else:
        val = normalize('NFKD', row['DS_MARCA_DETALHE']).encode('ASCII','ignore').decode('ASCII')
    return val
```
