```python
# aprendemos um pouco sobre o 'LabelEncoder()' e como converter dados categóricos em dados numéricos
# vamos visualizar como estão as colunas que são formadas por String
# lá atrás usamos a sintaze 'sales_data.dtypes' lembra? 
print("Supplies Subgroup' : ",sales_data['Supplies Subgroup'].unique())
print("Region : ",sales_data['Region'].unique())
print("Route To Market : ",sales_data['Route To Market'].unique())
print("Opportunity Result : ",sales_data['Opportunity Result'].unique())
print("Competitor Type : ",sales_data['Competitor Type'].unique())
print("'Supplies Group : ",sales_data['Supplies Group'].unique())
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

#conversão das colunas categóricas em numéricas
sales_data['Supplies Subgroup'] = le.fit_transform(sales_data['Supplies Subgroup'])
sales_data['Region'] = le.fit_transform(sales_data['Region'])
sales_data['Route To Market'] = le.fit_transform(sales_data['Route To Market'])
sales_data['Opportunity Result'] = le.fit_transform(sales_data['Opportunity Result'])
sales_data['Competitor Type'] = le.fit_transform(sales_data['Competitor Type'])
sales_data['Supplies Group'] = le.fit_transform(sales_data['Supplies Group'])

sales_data.head()
```

Agora <b> SIM </b>, avançamos bem! Com todos os dados já convertidos podemos em breve a construção do <b> Modelo Preditivo </b>

<b> Conjunto de Treinamento e Conjunto de Testes </b><p>
Um algoritmo de aprendizado de máquina precisa ser treinado em um conjunto de dados para aprender as relações entre diferentes recursos e como esses recursos afetam a variável de destino. Para isso, precisamos dividir todo o <b>conjunto de dados em dois conjuntos</b>. Um é o conjunto de <b>treinamento</b> no qual vamos treinar nosso algoritmo para construir um modelo.<p> O outro é o conjunto de <b>testes</b> no qual testaremos nosso modelo para ver quão precisas são suas previsões.


```python
#selecionando as colunas 'Opportunity Number' e 'Opportunity Result'
#não precisaremos das colunas citadas, pois a 1ª refere-se a um identificador excluisivo para cada registros e 
#e a outra é o nosso alvo! O que buscamos logo, não faz sentido possuir as mesmas
cols = [col for col in sales_data.columns if col not in ['Opportunity Number','Opportunity Result']]

data = sales_data[cols]
target = sales_data['Opportunity Result']
data.head(n=2)
```

Estamos todos prontos para definir nossos recursos e segmentar em dois quadros de dados separados. Em seguida, vamos dividir as dataframes 'data' e target' em conjuntos de formação e conjuntos de testes. Ao dividir o conjunto de dados, manteremos <b>30% dos dados como dados de teste</b> e os <b> 70% restantes como dados de treinamento</b>. Porém, fique a vontade para usar o modelo mais a sua escolha, usa-se muito o príncipio <b> '80/20'</b>, no qual são mantidos '80%' com dados de treinamento e os outros 20%' como dados de teste.

```python
#criação do 'data_train' e do 'data_test'
from sklearn.model_selection import train_test_split

data_train, data_test, target_train, target_test = train_test_split(data,target, test_size = 0.30, random_state = 10)
```

Para a escolha do melhor algoritmo use o <b> MAP </b> como referência<p>

O algoritmo que vamos usar para os nossos dados de vendas será o <b> [Gaussian Naive Bayes](http://scikit-learn.org/stable/modules/naive_bayes.html)</b> !?!?!<p>

**Acurácia (Accuracy)**, determina a proporção de casos que foram corretamente previstos

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

gnb = GaussianNB()

pred = gnb.fit(data_train, target_train).predict(data_test)

print("Naive-Bayes accuracy : ",accuracy_score(target_test, pred, normalize = True))
```
